# Gráfica de $f(x)=ln(x)$ con Sympy

Utilizo la librería sympy de python para realizar la gráfica de la serie de Taylor de orden 4 de la funcion $f(x)=ln(x)$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Se inicializan las variables $x_0$ y $n$ que indican el punto inicial de la serie y el orden de la serie, respectivamente.

```python
x0=1
n=5
```

A x se le asigna el valor de la variable $x$ con sympy

```python
x=sp.Symbol('x')
```

Se utiliza la función de sympy series() para obtener la serie de Taylor de grado 4 de la funcion $ln(x)$

```python
f=sp.log(x).series(x,x0,n).removeO()
f
```

La serie de Taylor obtenida en $f$ se convierte a una función manejable para python, la cual, pueda ser evaluada

```python
f_num=sp.lambdify(x,f,'numpy')
```

Se indica el rango de los valores de x en la gráfica. Además se obtienen los valores de la funcion logaritmo y la serie de taylor

```python
rangox=np.linspace(0.1,3,1000)
logy=np.log(rangox)
logt=f_num(rangox)
```

Se realiza la gráfica de las funciones

```python
plt.plot(rangox,logy,rangox,logt)
plt.legend(['Ln(x)','Serie de Taylor orden 4'])
plt.title('Serie de Taylor de $f(x)=ln(x)$')
plt.grid()
```
