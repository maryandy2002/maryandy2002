```python
import math as mt
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definici贸n de funciones de Newton-Cotes cerradas

def regla_trapezoidal(a, b, f):
    h = b - a
    return (h / 2) * (f(a) + f(b))

def regla_simpson(a, b, f):
    h = (b - a) / 2
    return (h / 3) * (f(a) + 4 * f((a + b) / 2) + f(b))

def regla_simpson_3_8(a, b, f):
    h = (b - a) / 3
    return (3 * h / 8) * (f(a) + 3 * f((2 * a + b) / 3) + 3 * f((a + 2 * b) / 3) + f(b))

def regla_boole(a, b, f):
    h = (b - a) / 4
    return (2 * h / 45) * (7 * f(a) + 32 * f((a + b) / 2) + 12 * f((a + 3 * b) / 4) + 32 * f((3 * a + b) / 4) + 7 * f(b))

# Ejemplo de uso

x = sp.symbols('x')
f = 1 / (1 + 25 * x**2)
f_lambdified = sp.lambdify(x, f, 'numpy')

a, b = -1, 1

# Evaluaci贸n de las reglas cerradas
resultado_trapezoidal = regla_trapezoidal(a, b, f_lambdified)
resultado_simpson = regla_simpson(a, b, f_lambdified)
resultado_simpson_3_8 = regla_simpson_3_8(a, b, f_lambdified)
resultado_boole = regla_boole(a, b, f_lambdified)

print(f"Resultado Trapezoidal: {resultado_trapezoidal}")
print(f"Resultado Simpson: {resultado_simpson}")
print(f"Resultado Simpson 3/8: {resultado_simpson_3_8}")
print(f"Resultado Boole: {resultado_boole}")

# Graficar la funci贸n y los resultados

xx = np.linspace(a, b, 1000)
yy = f_lambdified(xx)

plt.plot(xx, yy, label='f(x)')
plt.title('Funci贸n f(x) y Resultados de Newton-Cotes')
plt.legend()
plt.grid(True)
plt.show()
```
