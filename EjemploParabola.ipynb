# Nueva sección
<font color=red>Gráfica de una Parábola $y=x^2$

```python
import numpy as np
import matplotlib.pyplot as plt
#Grafica de la función cuadrática

# Se define el rango
x = np.linspace(-10, 10, 400)

# Se define la función de la parábola
y = x**2

# Graficar la parábola
plt.plot(x, y, label=r'$y = x^2$')
plt.title("Gráfico de una parábola", color='purple')

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.show()
```
